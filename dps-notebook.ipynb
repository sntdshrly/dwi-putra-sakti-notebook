# Dataset

```python
# Path yang akan digunakan
path = "/kaggle/input/dps-nomor1/nomor1.csv"
path_2 = "/kaggle/input/dps-nomor2/nomor2.csv"
path_3 = "/kaggle/input/dps-nomor3/nomor3.csv"
path_4 = "/kaggle/input/dps-nomor4/nomor4.csv"
path_7 = "/kaggle/input/dps-nomor7/nomor7.csv"
path_8 = "/kaggle/input/dps-nomor8/nomor8.csv"
path_9 = "/kaggle/input/dps-nomor9/nomor9.csv"
path_10 = "/kaggle/input/dps-nomor10/nomor10.csv"
```

# Nomor 1

```python
# Membaca dataset dari csv
df = pd.read_csv(path)

# Membagi data di kolom "Nama_Artikel" berdasarkan koma dan memindahkannya ke kolom berikutnya
df[['Nama_Artikel', 'Nama_Toko', 'Bulan_Tahun', 'Total_Qty_PCS']] = df['Nama_Artikel'].str.split(',', expand=True)

# Menghapus tanda kutip ganda ("") yang mungkin masih ada pada data
df['Nama_Artikel'] = df['Nama_Artikel'].str.strip('"')
df['Nama_Toko'] = df['Nama_Toko'].str.strip('"')

# Menampilkan lima data pertama
df
```

```python
# Ubah tipe data 'object' ke dalam 'integer'
df['Total_Qty_PCS'] = df['Total_Qty_PCS'].astype(int)
```

```python
# Ubah 'Bulan_Tahun' ke dalam nama bulan
df['Bulan_Tahun'] = pd.to_datetime(df['Bulan_Tahun'], format='%Y%m').dt.strftime('%B')
df
```

```python
# 'custom_order' : digunakan untuk patokan mengurutkan data
custom_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
df['Bulan_Tahun'] = pd.Categorical(df['Bulan_Tahun'], categories=custom_order, ordered=True)
```

```python
# Group nama toko, bulan_tahun, dan total qty pcs
grouped_data_months = df.groupby(['Nama_Toko', 'Bulan_Tahun'])['Total_Qty_PCS'].sum().unstack().fillna(0)
grouped_data_months
```

```python
# Plot bar chart
ax = grouped_data_months.plot(kind='bar', figsize=(12, 6))
plt.title('Total Qty PCS per Nama Toko per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Total Qty PCS')
plt.xticks(rotation=0)  # Putar x-axis labels ke 0 derajat

# Tampilkan plot
plt.show()
```

# Nomor 2

```python
# Membaca dataset dari csv
df = pd.read_csv(path_2)

# Menampilkan lima data pertama
df
```

```python
# Konversi kolom Bulan_Tahun ke format datetime
df['Bulan_Tahun'] = df['Bulan_Tahun'].apply(lambda x: datetime.strptime(x, '%B %Y'))
df
```

```python
# Filter data berdasarkan Nama_Toko tertentu (Karena jika dilihat dari dataset kebetulan hanya ada 1 toko)
nama_toko_terpilih = "GALERIA BIP BANDUNG"
df_filtered = df[df['Nama_Toko'] == nama_toko_terpilih]

# Ubah kolom Bulan_Tahun menjadi datetime
df_filtered['Bulan_Tahun'] = pd.to_datetime(df_filtered['Bulan_Tahun'], format='%Y%m')

# Urutkan DataFrame berdasarkan tanggal (Bulan_Tahun)
df_filtered = df_filtered.sort_values(by='Bulan_Tahun')

# Plot tren chart
plt.figure(figsize=(10, 6))
plt.plot(df_filtered['Bulan_Tahun'], df_filtered['Total_Penjualan_Per_Bulan'], marker='o', linestyle='-')
plt.title('Tren Penjualan Per Bulan')
plt.xlabel('Periode')
plt.ylabel('Total Penjualan Per Bulan')
plt.grid(True)
plt.xticks(rotation=0) 
plt.show()
```

# Nomor 3

```python
# Membaca dataset dari csv
df = pd.read_csv(path_3)

# Menampilkan lima data pertama
df
```

```python
# Melakukan pivot untuk mengubah data menjadi format yang cocok untuk stacked column plot
pivot_df = df.pivot(index='Bulan_Tahun', columns='Nama_Brand', values='Total_Qty_PCS')
pivot_df
```

```python
# Membuat stacked column plot
ax = pivot_df.plot(kind='bar', stacked=True, figsize=(10, 6))

# Menambahkan label dan judul
plt.xlabel('Periode')
plt.ylabel('Total Qty PCS')
plt.title('Total PCS Penjualan per Brand')
plt.legend(title='Nama Brand')
plt.xticks(rotation=0)
plt.show()
```
